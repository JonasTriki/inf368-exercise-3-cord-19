```python
# Imports
import numpy as np
rng_seed = 368
np.random.seed(rng_seed)
from mongoengine import connect, Document, StringField, IntField, MapField
from tqdm.auto import tqdm
```

```python
# Connect to db
connect('coordle')
```

```python
# Define schema for db
class Article(Document):
    cord_uid = StringField(required=True, primary_key=True)
    word_frequencies = MapField(field=IntField(), required=True)
```

```python
# Delete all articles
Article.objects().delete()
```

```python
# Saving some articles to db
for i in tqdm(range(500), unit='article'):
    cord_uid = f'lalala_{i + 1}'
    article = Article(cord_uid=cord_uid)
    for j in range(800):
        article.word_frequencies[f'word_{j + 1}'] = int(np.random.random() * 1000)
    article.save()
```

```python
# Looking up specific cord_uid
Article.objects(cord_uid='lalala_1')[0].word_frequencies
```
